## LeNet Implementation with Keras and MNIST dataset

### 1. Loading MNIST Database

```python
from keras.datasets import mnist

#pre-shuffled datasets
(X_train,y_train),(X_test,y_test)=mnist.load_data()


print(len(X_train))
print(len(X_test))



```

## 2. Visualizing Images

```python
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import numpy as np


#visualize the first six training images
fig=plt.figure(figsize=(25,25))

for i in range(6):
    ax=fig.add_subplot(1,6,i+1,xticks=[],yticks=[])
    ax.imshow(X_train[i],cmap='gray',vmin=0,vmax=255)
    ax.set_title(str(y_train[i]))
```

## 3. Detailed Analysis of an Image

```python
def visualize_input(img, ax):
    ax.imshow(img, cmap='gray')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],2)), xy=(y,x),
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')

fig = plt.figure(figsize = (12,12)) 
ax = fig.add_subplot(111)
visualize_input(X_train[3], ax)
```

## 4. Preprocessing Input Images

```python
#normalize data to accelerate learning
#Rescaling images by dividing evvery pixel in every image by 255
mean=np.mean(X_train)
print(mean)

#Standard deviation
std=np.std(X_train)


X_train=(X_train-mean)/(std+1e-7)
X_test=(X_test-mean)/(std+1e-7)

```
